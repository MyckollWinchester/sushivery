```python
import random
import faker
from datetime import datetime, timedelta

# Inicializar Faker
fake = faker.Faker('es_CL')

def generate_user():
    """Genera un usuario aleatorio."""
    id_number = f"{random.randint(1, 23)}.{random.randint(100, 999)}.{random.randint(100, 999)}-{random.randint(0, 9)}"
    name = fake.name()
    address = fake.address().replace("\n", ", ")
    comuna = fake.city()
    region = fake.region()
    birth = fake.date_of_birth(minimum_age=18, maximum_age=80).strftime('%Y-%m-%d')
    sexo = random.choice(["Masculino", "Femenino"])
    email = fake.email()
    phone = fake.phone_number()

    return {
        "id": id_number,
        "name": name,
        "address": address,
        "comuna": comuna,
        "region": region,
        "birth": birth,
        "sexo": sexo,
        "email": email,
        "phone": phone
    }

def generate_users(count=10):
    """Genera una lista de usuarios."""
    return [generate_user() for _ in range(count)]

# Generar 10 usuarios de ejemplo
users = generate_users(21)

# Mostrar los usuarios generados
import json
print(json.dumps(users, indent=4, ensure_ascii=False))
```

```python
import os

def export_to_data_ts(users, file_path):
    """Exporta la lista de usuarios al archivo data.ts."""
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write('import type { User } from \"@/types/user\"\n\n')
        f.write('export const users: User[] = [\n')
        for user in users:
            f.write(f'    {json.dumps(user, ensure_ascii=False)},\n')
        f.write(']\n')

# Exportar los usuarios generados a data.ts
output_path = os.path.join('c:\\Users\\gabol\\OneDrive\\Documents\\GitHub\\sushivery\\frontend\\components\\usertable', 'data.ts')
export_to_data_ts(users, output_path)
print(f"Archivo generado en: {output_path}")
```
